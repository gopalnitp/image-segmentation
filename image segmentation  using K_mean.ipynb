```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("highway-code-rule-007 (1).jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
img_F=np.reshape(img,(-1,3))
print img_F.shape
```

```python
from sklearn.cluster import KMeans
```

```python
color_dom=10
A=KMeans(color_dom)
A.fit(img_F)
```

```python
All_center=A.cluster_centers_
```

```python
All_center
```

```python
All_center=np.array(All_center,dtype="int32")
```

```python
All_center
```

```python
i=1
plt.figure(0,(10,2))
col=[]
for ix in All_center:
    plt.subplot(1,10,i)
    plt.axis("off")
    i+=1
    col.append(ix)
    b=np.zeros((100,100,3),dtype="int32")
    b[:,:,:]=ix
    plt.imshow(b)
plt.show()

```

```python
img_F.shape
```

```python
prdict_img=np.zeros(img_F.shape,dtype="int32")
```

```python
A.labels_
```

```python
for ix in range(prdict_img.shape[0]):
    prdict_img[ix]=col[A.labels_[ix]]
    
    
prdict_img=prdict_img.reshape(img.shape)
    
```

```python
print("Predict img")
plt.figure(0,(100,200))
plt.imshow(prdict_img)
plt.show()
```
